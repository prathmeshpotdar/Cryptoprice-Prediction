```python
import numpy as np
import pandas as pd
from keras.models import Model
from keras.layers import Input, Dense, Conv1D, Activation, GlobalAveragePooling1D, concatenate
from tensorflow.keras import layers


def create_rolling_windows(data, window_size):
    sequences = []
    for i in range(len(data) - window_size):
        sequence = data[i:i + window_size][column_names].values
        sequences.append(sequence)
    return np.array(sequences)

from keras import backend as K
def root_mean_squared_error(y_true, y_pred):
    return K.sqrt(K.mean(K.square(y_pred - y_true)))
    
from tensorflow.keras import backend as K

def mean_absolute_percentage_error(y_true, y_pred):
    y_true = K.cast(y_true, dtype=K.floatx())
    y_pred = K.cast(y_pred, dtype=K.floatx())
    return K.mean(K.abs((y_true - y_pred) / y_true)) * 100

from keras.optimizers import Adam


# Initial learning rate
initial_lr = 0.001

# Create an instance of the Adam optimizer with the initial learning rate
optimizer = Adam(lr=initial_lr)






# Hyperparameters
num_classes = 1   # Predicting a single value (Adjusted Close)
batch_size = 64
epochs = 800





window_size=best_window_size


# Create rolling windows of data
data_windowed1 = create_rolling_windows(data1, window_size)
target1 = data1['BTCPrice'].values[window_size:]

# Create rolling windows of data
data_windowed2 = create_rolling_windows(data2, window_size)
target2 = data2['BTCPrice'].values[window_size:]


# Split data into training and testing sets
train_size = int(1 * len(data_windowed1))
train_X, test_X = data_windowed1[:train_size], data_windowed1[train_size:]
train_y, test_y = target1[:train_size], target1[train_size:]
# Split data into training and testing sets
train_size = int(0 * len(data_windowed2))
train_X1, test_X = data_windowed2[:train_size], data_windowed2[train_size:]
train_y1, test_y = target2[:train_size], target2[train_size:]





# Build the MLP model with the current window size
input_layer = Input(shape=(window_size, l))
lstm_layer = LSTM(units=50, activation='relu',return_sequences=True)(input_layer)
lstm_layer = LSTM(units=50, activation='relu',return_sequences=False)(lstm_layer)
output_layer = Dense(num_classes)(lstm_layer)

model = Model(inputs=input_layer, outputs=output_layer)

# Compile the model with the RMSE loss function
model.compile(optimizer=optimizer, loss=root_mean_squared_error, metrics=[mean_absolute_percentage_error])

# Train the model on the training set and validate on the validation set
model.fit(train_X, train_y, epochs=epochs, batch_size=batch_size, validation_data=(test_X, test_y))

# Evaluate the model on the validation set
val_loss = model.evaluate(test_X, test_y)



# Step 4: Select the best window size
print(f"Validation Loss: {val_loss}")

```
